```python
import sys
import re
import os
import time
import random
import shutil
```

```python
base_path = './lofter'
save_path = './lofter_works'

dirpath = os.listdir(base_path)
```

```python
for i,ldir in enumerate(dirpath):
    works_path = os.path.join(base_path,ldir)
    works = os.listdir(works_path)
    print(ldir)
    for work in works:
        print("    "+os.path.join(works_path, work))
        if "works_" in work and '.html' in work:
            fh = open(os.path.join(works_path, work))
            work_str = fh.read()
            fh.close()
            if "" in work_str and 'userstuff' in work_str:
                shutil.copyfile(os.path.join(works_path, work), os.path.join(save_path, str(i+1)+"_"+work))
```
